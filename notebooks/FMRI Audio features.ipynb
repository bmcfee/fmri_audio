```python
import numpy as np
import librosa
import matplotlib.pyplot as plt
import prettyplotlib as ppl
%matplotlib inline
```

```python
def compute_audio_features(filename):
    
    # Load the input file with a default sample rate of 22050
    y, sr = librosa.load(filename)
    
    # Compute the harmonic and percussive components
    y_harm, y_perc = librosa.effects.hpss(y)
    
    # Compute onset strength, onset events and beat events
    # This uses the onset strength function of [McFee & Ellis, ICASSP 2005]
    onset_strength = librosa.onset.onset_strength(y=y_perc, 
                                                  sr=sr, 
                                                  hop_length=64, 
                                                  aggregate=np.median)
    
    onset_frames = librosa.onset.onset_detect(onset_envelope=onset_strength, 
                                              sr=sr, 
                                              hop_length=64)
    
    tempo, beat_frames = librosa.beat.beat_track(onset_envelope=onset_strength,
                                                 sr=sr,
                                                 hop_length=64)
    
    # Pad on a zero-marker for the beat frames
    beat_frames = np.unique(np.concatenate( ([0], beat_frames) ))
    
    # Convert frame numbers to time
    onset_times = librosa.frames_to_time(onset_frames, sr=sr, hop_length=64)
    beat_times = librosa.frames_to_time(beat_frames, sr=sr, hop_length=64)
    
    # And convert back to frames at a hop of 512, instead of 64
    onset_fr512 = librosa.time_to_frames(onset_times, sr=sr, hop_length=512)
    beat_fr512 = librosa.time_to_frames(beat_times, sr=sr, hop_length=512)
    
    # Compute mel spectrogram
    mel_spec = librosa.feature.melspectrogram(y=y, sr=sr).astype(np.float32)
    
    # Compute harmonic CQT
    cqt_spec = librosa.cqt(y=y_harm, sr=sr).astype(np.float32)
    
    # Compute chroma
    chroma = librosa.util.normalize(librosa.filters.cq_to_chroma(len(cqt_spec)).dot(cqt_spec)).astype(np.float32)
    
    # Beat synchronize
    mel_spec_beat_sync = librosa.feature.sync(mel_spec, beat_fr512)
    cqt_spec_beat_sync = librosa.feature.sync(cqt_spec, beat_fr512, aggregate=np.median)
    chr_beat_sync = librosa.feature.sync(chroma, beat_fr512, aggregate=np.median)
    
    return {'beat_times': beat_fr512,
            'tempo': tempo,
            'mel_spec': mel_spec,
            'mel_spec_beat_sync': mel_spec_beat_sync,
            'cqt_spec': cqt_spec,
            'cqt_spec_beat_sync': cqt_spec_beat_sync,
            'chroma': chroma,
            'chroma_beat_sync': chr_beat_sync}
```

```python
def display_features(features):
    
    plt.figure(figsize=(12,8))
    
    plt.subplot(311)
    librosa.display.specshow(librosa.logamplitude(features['mel_spec'], 
                                                  ref_power=np.max), 
                             y_axis='mel')
    plt.colorbar()
    plt.title('(log) Mel power spectrogram')
    
    plt.subplot(312)
    # Minimum frequency of our cqt is midi note 24 (C2)
    librosa.display.specshow(librosa.logamplitude(features['cqt_spec']**2.0, 
                                                  ref_power=np.max), 
                             y_axis='cqt_note',
                             fmin=librosa.midi_to_hz(24))
    plt.title('(log) Constant-Q power spectrogram')
    plt.colorbar()
    
    
    # Plot the chroma, and include a time scale on the bottom
    plt.subplot(313)
    librosa.display.specshow(features['chroma'], 
                             y_axis='chroma',
                             x_axis='time',
                             sr=22050,
                             hop_length=512)
    plt.title('Chroma')
    plt.colorbar()
    
    plt.tight_layout()
    
    
    plt.figure(figsize=(12,8))
    
    plt.subplot(311)
    librosa.display.specshow(librosa.logamplitude(features['mel_spec_beat_sync'], 
                                                  ref_power=np.max), 
                             y_axis='mel')
    plt.colorbar()
    plt.title('(log) Mel power spectrogram')
    
    plt.subplot(312)
    # Minimum frequency of our cqt is midi note 24 (C2)
    librosa.display.specshow(librosa.logamplitude(features['cqt_spec_beat_sync']**2.0, 
                                                  ref_power=np.max), 
                             y_axis='cqt_note',
                             fmin=librosa.midi_to_hz(24))
    
    plt.title('(log) Constant-Q power spectrogram')
    plt.colorbar()
    
    plt.subplot(313)
    librosa.display.specshow(features['chroma_beat_sync'], y_axis='chroma')
    plt.title('Chroma')
    plt.colorbar()
    
    plt.tight_layout()
```

```python
features = compute_audio_features('/home/bmcfee/Downloads/beethoven_sonata3Amajor_allegro_norm_mono.flac')
```

```python
display_features(features)
```
